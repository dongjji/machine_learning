```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/enrolment_1.csv')
df
```

```python
df['status'] = 'allowed'
info_tech_cond = (df['course name'] == 'information technology') & (df['year'] == 1)
commerce_cond = (df['course name'] == 'commerce') & (df['year'] == 4)
df.loc[info_tech_cond | commerce_cond, 'status'] = 'not allowed'
```

```python
count_stu = df['course name'].value_counts()
stu_upper_80 = list(count_stu[count_stu >= 80].index)
stu_upper_40 = list(count_stu[count_stu[count_stu >= 40] & count_stu[count_stu < 80]].index)
stu_upper_15 = list(count_stu[count_stu[count_stu >= 15] & count_stu[count_stu < 40]].index)
stu_upper_5 = list(count_stu[count_stu[count_stu >= 5] & count_stu[count_stu < 15]].index)
stu_less_5 = list(count_stu[count_stu <5].index)
```

```python
for course in stu_upper_80:
    df.loc[df['course name'] == course, 'room assignment'] = 'Auditorium'
    
for course in stu_upper_40:
        df.loc[df['course name'] == course, 'room assignment'] = 'Large room'
        
for course in stu_upper_15:
        df.loc[df['course name'] == course, 'room assignment'] = 'Medium room'        
        
for course in stu_upper_5:
        df.loc[df['course name'] == course, 'room assignment'] = 'Small room'

for course in stu_less_5:
        df.loc[df['course name'] == course, 'room assignment'] = 'not assigned'
```
