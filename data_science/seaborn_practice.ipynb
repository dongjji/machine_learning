## Kernel Density Estimation(KDE)

```python
df.head()
```

```python
df['Height']
```

```python
df['Height'].value_counts()
```

```python
df['Height'].value_counts().sort_index()
```

```python
df['Height'].value_counts().sort_index().plot()
```

```python
sns.kdeplot(df['Height'])
```

```python
sns.kdeplot(df['Weight'])
```

```python
df.plot(kind="scatter", x="Height", y="Weight")
```

```python
sns.kdeplot(df['Height'], bw_method=0.1)
```

```python
df.plot(kind="hist", y="Height", bins=15)
```

```python
sns.distplot(df['Height'], bins=15)
```

```python
df.plot(kind="box", y='Height')
```

```python
sns.violinplot(y=df['Height'])
```

```python
df.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.kdeplot(x=df['Height'], y=df['Weight'])
```

```python
sns.lmplot(data=df, x='Height', y='Weight')
```
