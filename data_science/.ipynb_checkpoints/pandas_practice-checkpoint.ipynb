###  값 고치기

```python
iphone_df2.loc[['iPhone 8']]
```

```python
iphone_df2.loc['iPhone 8', '메모리'] = '3GB'
iphone_df2.loc[['iPhone 8']]
```

```python
iphone_df2['제조사'] = 'Apple'
iphone_df2
```

```python
iphone_df2.loc['iPhone XR'] = ['1111-11-11', 4.9, '3GB', 'iOS 12.0.1', 'Yes', 'Apple']
iphone_df2
```

```python
iphone_df2.drop('iPhone XR', axis='index', inplace=False) # inplace=False하면 기존 df는 변경되지않음 다른변수에 저장할때 사용
```

```python
iphone_df2
```

```python
iphone_df2.drop('iPhone XR', axis='index', inplace=True)
iphone_df2
```

```python
iphone_df2.drop(['iPhone 7', 'iPhone 8', 'iPhone X'], axis='index', inplace=False)
```

```python
iphone_df2.rename(columns={'Face ID': 'FACE ID'})
```

```python
iphone_df2
```

```python
iphone_df2.rename(columns={'Face ID': 'FACE ID'}, inplace= True)
iphone_df2
```

```python
iphone_df2.index.name = '기종'
```

```python
iphone_df2
```

```python
iphone_df2['기종'] = iphone_df2.index
iphone_df2
```

```python
iphone_df2.set_index('제조사', inplace=True)
```
