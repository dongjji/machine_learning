```python
df = pd.read_csv('data/silicon_valley.csv')
```

```python
adobe_condition = (df['company'] == 'Adobe') & (df['count'] > 0) & (df['race'] == 'Overall_totals')
df_adobe = df.loc[adobe_condition, ['job_category', 'count']][0:5]
df_adobe.set_index('job_category', inplace=True)
```

```python
df_adobe
```

```python
df_adobe.plot(kind="pie", y='count')
```
